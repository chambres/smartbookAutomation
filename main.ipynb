````python
#vars
USERNAME = ""
PASSWORD = ""
connect_ed_module_url = "https://katyisd.instructure.com/courses/515146/modules/items/24647317" #yours will be different if you have a different teacher
wait_time = 10
````

````python
#open chrome, garden gnome
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementClickInterceptedException

ignored_exceptions=(NoSuchElementException,StaleElementReferenceException)

driver = webdriver.Chrome(r"chromedriver.exe")

def switch_to_first_tab():
    driver.switch_to.window(driver.window_handles[0])
    
switch_to_first_tab()

driver.get("https://katyisd.instructure.com/")

driver.find_elements(By.ID, "userNameInput")[0].send_keys(USERNAME + "@students.katyisd.org")
driver.find_elements(By.ID, "passwordInput")[0].send_keys(PASSWORD)
driver.find_element(By.ID, "submitButton").click()

try:
    WebDriverWait(driver, 300).until(
        EC.presence_of_element_located((By.ID, "dashboard"))
    )
finally:
    driver.get(connect_ed_module_url)
    try:
        WebDriverWait(driver, wait_time, ignored_exceptions=ignored_exceptions).until(
            EC.presence_of_element_located((By.CLASS_NAME, "btn"))
            )
    finally:
        driver.find_elements(By.CLASS_NAME, "btn")[0].click()
````

````python
switch_to_first_tab()

def click_confidence():
    highconfbutton = driver.find_element_by_css_selector("[aria-label='High Confidence']")

    try:
        WebDriverWait(driver, wait_time, ignored_exceptions=ignored_exceptions).until(
            lambda driver: "disabled" not in highconfbutton.get_attribute("outerHTML")
        )
    finally:

        highconfbutton.click()
        
        try:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            WebDriverWait(driver, wait_time, ignored_exceptions=ignored_exceptions).until(
                EC.presence_of_element_located((By.CLASS_NAME, "lr__action-label"))
                )
        finally:
            
            driver.find_element(By.CLASS_NAME, "lr__action-label").click()
            #print('clicked to questions')

            try:
                WebDriverWait(driver, wait_time, ignored_exceptions=ignored_exceptions).until(
                    EC.presence_of_element_located((By.CLASS_NAME, "btn.ngx-shared.btn-primary"))
                )
            finally:
                if(driver.find_element(By.CLASS_NAME, "btn.ngx-shared.btn-primary").text == "To Questions"):
                    driver.find_element(By.CLASS_NAME, "btn.ngx-shared.btn-primary").click()

                try:
                    WebDriverWait(driver, wait_time, ignored_exceptions=ignored_exceptions).until(
                        EC.presence_of_element_located((By.CLASS_NAME, "btn.ngx-shared.btn-primary"))
                    )
                finally:
                    if(driver.find_element(By.CLASS_NAME, "btn.ngx-shared.btn-primary").text == "Next Question"):
                        driver.find_element(By.CLASS_NAME, "btn.ngx-shared.btn-primary").click()
                    

            
         

def mcq():
    try:
        WebDriverWait(driver, 10, ignored_exceptions=ignored_exceptions).until(
            EC.presence_of_element_located((By.CLASS_NAME, "choice-row"))
        )
    finally:
        driver.find_elements(By.CLASS_NAME, "choice-row")[0].click()
        click_confidence()
def fill_in_the_blank():
    try:
        WebDriverWait(driver, wait_time, ignored_exceptions=ignored_exceptions).until(
            EC.presence_of_element_located((By.TAG_NAME, "input"))
            )
    finally:
        for i in driver.find_elements(By.CLASS_NAME, "prompt")[0].find_elements(By.TAG_NAME, "input"):
            i.send_keys('haha')

        click_confidence() 
def tof():
    try:
        WebDriverWait(driver, 10, ignored_exceptions=ignored_exceptions).until(
            EC.presence_of_element_located((By.CLASS_NAME, "ahe-ui-radio"))
        )
    finally:
        driver.find_elements(By.CLASS_NAME, "ahe-ui-radio")[0].click()
        click_confidence()
def forge():
    js = """
            var num = document.getElementsByClassName("summary-block__value")[1].textContent;
        document.getElementsByClassName("summary-block__value")[1].style = "color:green";
        var l = num.split(" ")
        function randomNumber(min, max){
            const r = Math.random()*(max-min) + min
            return Math.floor(r)
        }
        l[1] = parseInt(l[4]) - randomNumber(50, 75)
        document.getElementsByClassName("summary-block__value")[1].textContent = l.join(" ");
        """
    driver.execute_script(js);
    driver.save_screenshot("smartbook.png")
````

The Main Loop: (please read)
==
The following cell of code is the main loop. Should anything fail, and you find the cell has crashed due to some unknown error, restarting THIS cell would be your best course of action.

Now, begin the loop by clicking the play button next to `try:`
and watch your Chrome tab go wild.

````python
switch_to_first_tab()
try:
    while True:
        try:
            WebDriverWait(driver, wait_time, ignored_exceptions=ignored_exceptions).until(
                EC.presence_of_element_located((By.CLASS_NAME, "probe-header"))
            )
        except KeyboardInterrupt:
            break
        finally:
            try:
            #print(driver.find_element(By.CLASS_NAME, "probe-header").text, end="\r", flush=True)    
                if driver.find_element(By.CLASS_NAME, "probe-header").text == "Fill in the Blank Question":
                    fill_in_the_blank()
                elif driver.find_element(By.CLASS_NAME, "probe-header").text == "Multiple Select Question" or driver.find_element(By.CLASS_NAME, "probe-header").text == "Multiple Choice Question":
                    mcq()
                elif driver.find_element(By.CLASS_NAME, "probe-header").text == "True or False Question":
                    tof()
            except KeyboardInterrupt:
                break
            except NoSuchElementException:
                continue
            except StaleElementReferenceException:
                continue
except KeyboardInterrupt:
    print("interrupted!")
````
